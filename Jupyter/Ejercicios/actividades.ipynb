<h4>1. Generar dos vectores aleatorios de valores enteros (valores de 1 a 10). Los vectores solamente deben tener 5 elementos y se deben sumar sus elementos para obtener un único vector.<h/>

```python
import numpy as np
```

```python
v1 = np.random.randint(1,10,5)
v1
```

```python
v2 = np.random.randint(1,10,5)
v2
```

```python
v3 = v1 + v2
v3
```

<h4>2. Utilizar la biblioteca de Matplotlib para Graficar el vector generado anteriormente<h/>

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(v3, marker = 'o', linestyle = "-", color = "red", label = "Vector")
plt.xlabel("Indice")
plt.ylabel("Valor")

plt.xticks(np.arange(0,5))
plt.yticks(np.arange(0,13))
plt.legend()
```

<h4>1. Generar un vector aleatorio que incluya números negativos desde el -10 hasta el 10 positivo. Este vector debe tener 25 elementos. Deben ser números flotantes.<h/>

```python
vector = np.random.uniform(-10,10,25)
vector
```

<h4>2. Transformar el vector en una matriz cuadrada de 5 x 5. Obtener el promedio de la matriz.<h/>

```python
matriz = vector.reshape(5,5)
matriz
```

<h4>3. Generar un mapa de calor a partir de la matriz generada anteriormente.<h/>

```python
#Crear una figura y un eje
fig, grafico = plt.subplots()

#Mostrar la imagen el mapa de calor de la matriz
cgrafico = grafico.imshow(matriz, cmap='viridis', interpolation='nearest')

#Añadir una barra de calor
fig.colorbar(cgrafico)

#Añadir título y etiquetas a los ejes
grafico.set_title("Mapa de Calor de una Matriz de 5x5")
grafico.set_xlabel("Columnas")
grafico.set_ylabel("Filas")

#Agregando la matriz cuadrada de 5x5
x = np.random.random((5,5))

#Mostrar el gráfico
grafico.imshow(x)
```

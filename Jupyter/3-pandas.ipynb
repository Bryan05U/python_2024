#### **FUNCIÃN PARA VERIFICAR SI EL GENERO PUZZLE ESTÃ PRESENTE EN LA LISTA**

```python
def filtro_puzzle(genres):
    return 'Puzzle' in genres
# Se aplica el filtro utilizando apply
datosPuzzle = dataframe[dataframe['Genres'].apply(filtro_puzzle)]
```

#### **TRANSFORMACIÃN DE DATASET: MUESTRA 2 COLUMNAS DE DATOS (SOLO JUEGOS PUZZLE)**

```python
# Muestra el dataset solamente incluyendo las columnas Title y Rating
nombresPuzzle = datosPuzzle[['Title','Rating']]
nombresPuzzle.head(20)
```

#### **MOSTRANDO LOS PRIMEROS 20 VIDEOJUEGOS DEL RATING CON SU TÃTULO CORRESPONDIENTE**

```python
datasetPuzzle = nombresPuzzle.head(20)
datasetPuzzle.set_index('Title', inplace=True)

grafico = datasetPuzzle['Rating'].plot(x=None, kind='bar')

# Agregando etiquetas de rating en las barras
for i, rating in enumerate(datasetPuzzle['Rating']):
    grafico.annotate(str(rating), xy=(i, rating), ha='center', va='bottom')

plt.show()
```
